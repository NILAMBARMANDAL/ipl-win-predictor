```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
matches = pd.read_csv("data/matches.csv")
deliveries = pd.read_csv("data/deliveries.csv")
```

```python
print(matches.shape)
print(deliveries.shape)
```

```python
matches['season'].value_counts().sort_index().plot(kind='bar')
plt.title("matches per season")
plt.show()
```

```python
matches['winner'].value_counts().head(10).plot(kind='bar')
plt.title("Top Winning Teams")
plt.show()
```

```python
deliveries.head()
```

```python
top_players = deliveries.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)

top_players.plot(kind='bar')
plt.title("Top Batsmen")
plt.show()
```

```python
deliveries.columns
```

```python
toss_win_match_win = matches[matches['toss_winner'] == matches['winner']]

# percentage
percentage = (len(toss_win_match_win) / len(matches)) * 100

print(f"Teams winning toss also won match: {percentage:.2f}%")
```

```python
import seaborn as sns

sns.countplot(x=(matches['toss_winner'] == matches['winner']))
plt.xticks([0,1], ['Lost Match', 'Won Match'])
plt.title("Impact of Toss on Match Result")
plt.show()
```

```python
matches['result'].value_counts()
```

```python
matches['result'].value_counts().plot(kind='bar')

plt.title("Batting First vs Chasing Wins")
plt.xlabel("Result Type")
plt.ylabel("Number of Matches")
plt.show()
```
